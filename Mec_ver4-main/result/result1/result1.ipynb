```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
import pandas as pd
plt.rcParams['figure.figsize'] = [12, 4]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.tight_layout()
```

```python
a = 40
b = 100
```

```python
folder_name = "DQN1" + '/'
folder_name2 = "DQN2_no_energy" + '/'

data_delay = pd.read_csv(folder_name + "delay.csv").values[a:b]
number = [i + a + 1 for i in range(data_delay.shape[0])]
plt.plot(number, data_delay, color="red", label = "Delay")
# data_delay = pd.read_csv(folder_name2 + "delay.csv").values[a:b]
# plt.plot(number, data_delay, color="yellow", label = "Delay")
plt.title("Delay")
plt.show()
plt.clf()

data_energy = pd.read_csv(folder_name + "energy.csv").values[a:b]
plt.plot(number, data_energy, color="red", label = "Energy")
# data_energy = pd.read_csv(folder_name2 + "energy.csv").values[a:b]
# plt.plot(number, data_energy, color="yellow", label = "Reward")
plt.title("Energy")
plt.show()
plt.clf()

data_reward = pd.read_csv(folder_name + "reward.csv").values[a:b]
plt.plot(number, data_reward, color="red", label = "Reward")
# data_reward = pd.read_csv(folder_name2 + "reward.csv").values[a:b]
# plt.plot(number, data_reward, color="yellow", label = "Reward")
plt.title("Reward")
plt.show()
plt.clf()
```

```python
folder_name = "DQL_without_energy" + '/'

data_delay = pd.read_csv(folder_name + "DQL_delay.csv", header=None).values[a:b]
number = [i + a + 1 for i in range(data_delay.shape[0])]
plt.plot(number, data_delay, color="yellow", label = "Delay")
plt.title("Delay")
plt.show()
plt.clf()

data_energy = pd.read_csv(folder_name + "DQL_energy.csv", header=None).values[a:b]
plt.plot(number, data_energy, color="yellow", label = "Energy")
plt.title("Energy")
plt.show()
plt.clf()

data_reward = pd.read_csv(folder_name + "DQL_reward.csv", header=None).values[a:b]
plt.plot(number, data_reward, color="yellow", label = "Reward")
plt.title("Reward")
plt.show()
plt.clf()
```
