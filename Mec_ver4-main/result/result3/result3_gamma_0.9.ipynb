```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
import pandas as pd
plt.rcParams['figure.figsize'] = [12, 4]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.tight_layout()
```

```python
a = 0
b = 100
```

```python
metric_file = "drop_task.csv"
folder_name1 = "DQN1" + '/'
folder_name2 = "DDQN1" + '/'
folder_name3 = "DuelingDQN1" + '/'
folder_name4 = "DoubleDuelingDQN1" + '/'
folder_name5 = "DQN1_no_energy" + '/'
folder_name6 = "DDQN1_no_energy" + '/'
folder_name7 = "DuelingDQN1_no_energy" + '/'
folder_name8 = "DQN2" + '/'

data_delay = pd.read_csv(folder_name1 + metric_file).values[a:b]
number = [i + a + 1 for i in range(data_delay.shape[0])]
plt.plot(number, data_delay, color="red", label = "DQN")
data_delay = pd.read_csv(folder_name2 + metric_file).values[a:b]
plt.plot(number, data_delay, color="yellow", label = "Double DQN")
data_delay = pd.read_csv(folder_name3 + metric_file).values[a:b]
plt.plot(number, data_delay, color="green", label = "Dueling DQN")

line = '--'
data_delay = pd.read_csv(folder_name5 + metric_file).values[a:b]
plt.plot(number, data_delay, color="red", label = "DQN no energy", linestyle = line)
data_delay = pd.read_csv(folder_name6 + metric_file).values[a:b]
plt.plot(number, data_delay, color="yellow", label = "Double DQN no energy", linestyle = line)
data_delay = pd.read_csv(folder_name7 + metric_file).values[a:b]
plt.plot(number, data_delay, color="green", label = "Dueling DQN no energy", linestyle = line)

plt.legend()
plt.title("Task drop")
plt.show()
plt.clf()
```
