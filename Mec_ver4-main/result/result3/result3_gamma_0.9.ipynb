```python
import matplotlib.pyplot as plt
plt.clf()
colors = ["crimson", "purple", "gold"]
f = lambda m,c: plt.plot([],[],marker=m, color=c, ls="none")[0]
handles = [f("s", colors[i]) for i in range(3)]
labels = colors
legend = plt.legend(handles, labels, loc=3, framealpha=1, frameon=False)

def export_legend(legend, filename="legend.png"):
    fig  = legend.figure
    fig.canvas.draw()
    bbox  = legend.get_window_extent().transformed(fig.dpi_scale_trans.inverted())
    fig.savefig(filename, dpi="figure", bbox_inches=bbox)

export_legend(legend)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('default')
import pandas as pd

plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower

def allocation_statistics(folder_name):
    Cov = pd.read_csv(folder_name + "server_allocation.csv", sep=',')
    Cov.columns = ["BS", "VS1", "VS2", "VS3", "VS4", "VS5"]
    #Cov.rename(columns={"server": "a", "bus1": "b", "bus2": "c", "bus3": "d", "bus4": "e", "bus5":"f"})
    columns_titles = ["BS", "VS1", "VS2", "VS3", "VS4", "VS5"]
    df = Cov.divide(Cov.sum(axis=1), axis=0)
    df = df.reindex(columns=columns_titles)

    df.plot.area(legend = 'reverse')
```

```python
allocation_statistics(folder_name1)
allocation_statistics(folder_name2)
allocation_statistics(folder_name3)
allocation_statistics(folder_name5)
allocation_statistics(folder_name6)
allocation_statistics(folder_name7)
```
